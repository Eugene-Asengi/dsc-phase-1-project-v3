```python
Aviation_data = Aviation_data_copy.copy()
```

```python
Aviation_data.drop(columns=['Schedule', 'FAR.Description', 'Latitude', 'Longitude', 'Location', 'Air.carrier', 'Airport.Name'], inplace = True)
```

```python
print(Aviation_data.isna().mean()*100)
print(Aviation_data.shape)
```

```python
Aviation_data.describe()
```

```python
# Replacing missing values with the median
Aviation_data_columns = ['Number.of.Engines', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
Aviation_data[Aviation_data_columns] = Aviation_data[Aviation_data_columns].fillna(Aviation_data[Aviation_data_columns].median())
```

```python
Aviation_data.isna().sum()
```

```python
# Replacement of all missing categorical values
Aviation_data.fillna('Missing', inplace= True)
```

```python
Aviation_data.isna().sum()
```

```python
# Dropping duplicates
Aviation_data.drop_duplicates()
Aviation_data
```

**Aggregation methods and .groupby()**  
In this section, we will explore the different aggregation methods and how to use the .groupby()
This helps in understanding different relationships between what causes accidents and what is the most aircraft prone to failure. Grouping the data using the groupby() method is highly beneficial especially for categorical variables that may provide insight into how different factors influence injuries, fatalities and other metrics.

```python
Aviation_data.columns
```

```python
Aviation_data
```

```python
# Grouping by Aircraft category and calculate mean number of accidents
df = Aviation_data.groupby('Aircraft.Category').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
df = Aviation_data[Aviation_data['Aircraft.Category'] == 'Airplane'].groupby('Make').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
df = Aviation_data.groupby('Country').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
df = Aviation_data.groupby('Model').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
df=Aviation_data.groupby('Engine.Type').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
df = Aviation_data.groupby('Make').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```
