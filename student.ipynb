**Aggregation methods and .groupby()**  
In this section, we will explore the different aggregation methods and how to use the .groupby()
This helps in understanding different relationships between what causes accidents and what is the most aircraft prone to failure. Grouping the data using the groupby() method is highly beneficial especially for categorical variables that may provide insight into how different factors influence injuries, fatalities and other metrics.

```python
Aviation_data.columns
```

```python
# Grouping by Aircraft category and calculate mean number of accidents
df = Aviation_data.groupby('Aircraft.Category').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
# Grouping by Aircraft category, total injuries and the make of Airplanes
df = Aviation_data[Aviation_data['Aircraft.Category'] == 'Airplane'].groupby('Make').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```

```python
# Grouping by country and total injuries
df = Aviation_data.groupby('Country').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df.head(20)
```

```python
#Grouping aircraft models and total injuries
df = Aviation_data.groupby('Model').agg({
    'Total.Fatal.Injuries': 'mean',
    'Total.Serious.Injuries': 'mean',
    'Total.Minor.Injuries': 'mean',
    'Total.Uninjured': 'mean'
}).reset_index()
df
```
